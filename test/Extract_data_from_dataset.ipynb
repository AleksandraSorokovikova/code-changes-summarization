```python
import re
def remove_comments_with_whitespace(code: str) -> str:
  code_no_single_line_comments = re.sub(r'#.*', '', code)
  code_no_comments = re.sub(r'""".*?"""|\'\'\'.*?\'\'\'', '', code_no_single_line_comments, flags=re.DOTALL)
  return code_no_comments
print(remove_comments_with_whitespace(df['code'][12]))
```

```python
code_without_comments = df['code'].apply(remove_comments_with_whitespace)[:10]
```
